# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
Logistic_Crosstab = pd.crosstab(y_test, y_pred)
Logistic_Crosstab
```

```python
Accuracy  = (Logistic_Crosstab[0][0] + Logistic_Crosstab[1][1]) / len(y_test)
Precision = (Logistic_Crosstab[1][1]) /  (Logistic_Crosstab[1][1] + Logistic_Crosstab[1][0])
Recall    = (Logistic_Crosstab[1][1]) /  (Logistic_Crosstab[1][1] + Logistic_Crosstab[0][1])
F1Score   = 2*(Precision*Recall) / (Precision+Recall)

print("Accuracy {} %".format(100*Accuracy))
print("Precision {} %".format(100*Precision))
print("Recall {} %".format(100*Recall))
print("F1Score {} %".format(100*F1Score))
```

# Logistic Regression Model Finetuning

```python
# We will try to improve our model, by using a threshold:
```

```python
y_test2 = pd.DataFrame(y_test)
y_test2 = pd.DataFrame.reset_index(y_test2) 
probs = model.predict_proba(X_test)
prob = probs[:,1]
prob_df = pd.DataFrame(prob)
```

```python
threshold = [0.1, 0.2, 0.3, 0.33, 0.35, 0.38, 0.4, 0.5, 0.6, 0.7]
Recall = [1]
Precision = [1]
F1Score = [1]

for t in threshold:
    prob_df["prediction"] = np.where(prob_df[0]>= t, 1, 0)
    prob_df["actual"] = y_test2['Attrition']
    confusion_matrix = pd.crosstab(prob_df.prediction, prob_df.actual)
    TN = confusion_matrix[0][0]
    TP = confusion_matrix[1][1]
    FP = confusion_matrix[0][1]
    FN = confusion_matrix[1][0]
    rec = TP / (TP+FN)
    pre_1= 1-(TP/(TP+FP))
    pre = TP/(TP+FP)
    f1 = 2*(pre*rec) / (pre+rec)
    Recall.append(rec)
    Precision.append(pre_1)
    F1Score.append(f1)
    print("When Threshold equals " + str(t) + ", F1Score equals " + str(f1))

Recall.append(0)
Precision.append(0)
F1Score.append(0)
```

```python
prob_df = pd.DataFrame(prob)
threshold = 0.30
prob_df["prediction"] = np.where(prob_df[0]>= threshold, 1, 0)
prob_df['actual'] = y_test2['Attrition']
Logistic_Crosstab = pd.crosstab(prob_df['actual'], prob_df["prediction"])
Accuracy  = (Logistic_Crosstab[0][0] + Logistic_Crosstab[1][1]) / len(prob_df)
Precision = (Logistic_Crosstab[1][1]) /  (Logistic_Crosstab[1][1] + Logistic_Crosstab[1][0])
Recall    = (Logistic_Crosstab[1][1]) /  (Logistic_Crosstab[1][1] + Logistic_Crosstab[0][1])
F1Score   = 2*(Precision*Recall) / (Precision+Recall)
print("Accuracy {} %".format(100*Accuracy))
print("Precision {} %".format(100*Precision))
print("Recall {} %".format(100*Recall))
print("F1Score {} %".format(100*F1Score))
```

# Random Forest Model Finetuning

```python
estimators = [500, 1000, 2500, 5000, 10000, 15000, 20000]
for e in estimators:
    model = RandomForestClassifier(n_estimators = e)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    Forest_Crosstab = pd.crosstab(y_test, y_pred)
    Accuracy  = (Forest_Crosstab[0][0] + Forest_Crosstab[1][1]) / len(X_test)
    Precision = (Forest_Crosstab[1][1]) /  (Forest_Crosstab[1][1] + Forest_Crosstab[1][0])
    Recall    = (Forest_Crosstab[1][1]) /  (Forest_Crosstab[1][1] + Forest_Crosstab[0][1])
    F1Score   = 2*(Precision*Recall) / (Precision+Recall)
    print("Number of estimators: " + str(e) + ". F1Score: {} %".format(100*F1Score))
```

```python
# As we can see above, our model is stabilized after a certain amount of trees is used. 
# However, using that quantity does not affect the predicting capabilities of our model.
# Let's finally use a Random Tree Cross Validation Model
```
